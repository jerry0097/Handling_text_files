# Grouping and combined based on character

```python
df = script.sort_values("Name")
df
```

```python
df['Speech'] = df.groupby(['Name'])['Speech'].transform(lambda x : ' '.join(x))
df = df.drop_duplicates()
df
```

# Taking the speech, and getting unique words

```python
def to_split(txt):
    return txt.split()
```

```python
df['break'] = df['Speech'].apply(lambda x : to_split(x))
df
```

```python
df['break_set'] = df['break'].apply(lambda x : set(x))
```

```python
 df.to_csv(r"C:\Users\jason\Documents\Jerry\Studies\Jerry DSc\Python for DS- Project\conv2.csv", index = None)
```

```python
df['break_set'] = df['break'].apply(lambda x : list(x))
df
```

```python
df_1 = df.drop(columns=["Speech","break"], axis=1)
df_1
```

```python
df_1['break_set'] = df_1['break_set'].astype("string")
df_1.info()
```

```python
df_1 = df.drop(columns=["break_set","Speech","break"], axis=1)
df_2 = pd.DataFrame(df['break_set'].values.tolist())
df_1 = df_1.reset_index(drop=True)
df_2 = df_2.reset_index(drop=True)
clean = pd.concat([df_1, df_2], axis=1)
final = clean
final.to_csv(r"C:\Users\jason\Documents\Jerry\Studies\Jerry DSc\Python for DS- Project\speaker.csv")
```
